```python
import pandas as pd

data = pd.read_csv("Pilpres.csv")
data.head(5)
```

```python
data.info()
```

```python
data.drop_duplicates(subset ="full_text", keep = 'first', inplace= True)
data.info()
```

```python
df = pd.DataFrame(data['full_text'])
df.head(5)
```

CLEANING

```python
import re
import string
import nltk

def remove_URL(tweet):
    url = re.compile(r'https?://\S+|www\.\S+')
    return url.sub(r'',tweet)

def remove_html(tweet):
    html = re.compile(r'<.*?>')
    return html.sub(r'', tweet)

def remove_emoji(tweet):
    emoji_pattern = re.compile("["
        u"\U0001F600-\U0001F64F"
        u"\U0001F300-\U0001F5FF"
        u"\U0001F680-\U0001F6FF"
        u"\U0001F1E0-\U0001F1FF"
                          "]+", flags=re.UNICODE)
    return emoji_pattern.sub(r'', tweet)

def remove_numbers(tweet):
    tweet = re.sub(r'\d+', '', tweet)
    return tweet

def remove_symbols(tweet):
    tweet = re.sub(r'[^a-zA-Z0-9\s]', '', tweet)
    return tweet

df['cleasing'] = df['full_text'].apply(lambda x: remove_URL(x))
df['cleasing'] = df['cleasing'].apply(lambda x: remove_html(x))
df['cleasing'] = df['cleasing'].apply(lambda x: remove_emoji(x))
df['cleasing'] = df['cleasing'].apply(lambda x: remove_symbols(x))
df['cleasing'] = df['cleasing'].apply(lambda x: remove_numbers(x))

df.head(5)
```

CASEFOLDING

```python
def case_folding(text):
    if isinstance(text, str):
        lowercase_text = text.lower()
        return lowercase_text
    else:
        return text

df['case_folding'] = df['cleasing'].apply(case_folding)

df.head(5)
```

TOKENIZATION

```python
def tokenize(text):
  tokens = text.split()
  return tokens

df['tokenize'] = df['case_folding'].apply(tokenize)

df.head(5)
```

FILTERING/STOPWORD REMOVAL

```python
from nltk.corpus import stopwords
nltk.download('stopwords')
stop_words = stopwords.words('indonesian')
```

```python
def remove_stopwords(text):
    return [word for word in text if word not in stop_words]

df['Filtering/stopword removal'] = df['tokenize'].apply(lambda x: remove_stopwords(x))

df.head(5)
```

STEMMING DATA

```python
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory
from nltk.stem import PorterStemmer
from nltk.stem.snowball import SnowballStemmer
```

```python
factory = StemmerFactory()
stemmer = factory.create_stemmer()

def stem_text(text):
    return [word for word in text if stemmer.stem]

df['stemming_data'] = df['Filtering/stopword removal'].apply(lambda x: ' '.join(stem_text(x)))

df.head(5)
```
